2) 5 идентичных рекламных кампаний работали 24 дня по 6 часов в день, потрачено 
было 120 долларов. Сколько дней они работали бы на 216 долларов, если бы 9 
одинаковых кампаний работали бы по 8 часов в день?

6*24 = 144 часа работало 5 компаний

120 / 144 = 0.8(3) доллара в час

5 компаний зарабатывают 0.8(3) доллара в час

Х = 9*0.8(3) / 5 = 1.5 долларов в час

216 долларов / 1.5 = 144 часа работали 9 компаний по 8 часов в день

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
pd.options.display.float_format = '{:,.2f}'.format
```

```python
data = pd.read_excel('./data.xlsx')
```

```python
data = data.drop(columns = ['Unnamed: 5'],axis = 1) #удалим пустой столбец
```

Преобразуем строки с названием месяцев в отдельную колонку для удобства

```python
clients = []
months = ['май 2021', 'июнь 2021', 'июль 2021', 'август 2021', 'сентябрь 2021', 'октябрь 2021']
month = []
q = 0
for i in data['client_id'].iloc[1:]:
    if np.isnan(i) == False:
        month.append(months[q])
    else:
        clients.append(month)
        month = []
        q += 1
clients.append(month)
month_list = [i for m in clients
              for i in m]
```

```python
data = data.dropna(subset=['client_id']) #удалим ненужные строки с названием месяца
```

```python
data['month'] = month_list
data
```

```python
data.dtypes
```

Преобразуем поле с датой получения оригинала договора в более удобный формат для дальнейшей работы с ним

```python
data['receiving_date'] = data.receiving_date.apply(lambda x: x.strftime('%Y-%m-%d') if (x != '-' and pd.isnull(x) == False) else x)
```

### Задание

За каждую заключенную сделку менеджер получает бонус, который рассчитывается 
следующим образом.
1) За новые сделки менеджер получает 7 % от суммы, при условии, что статус 
оплаты «ОПЛАЧЕНО», а также имеется оригинал подписанного договора с 
клиентом (в рассматриваемом месяце).

2) За текущие сделки менеджер получает 5 % от суммы, если она больше 10 тыс., и 
3 % от суммы, если меньше. При этом статус оплаты может быть любым, кроме 
«ПРОСРОЧЕНО», а также необходимо наличие оригинала подписанного 
договора с клиентом (в рассматриваемом месяце).

Бонусы по сделкам, оригиналы для которых приходят позже рассматриваемого 
месяца, считаются остатком на следующий период, который выплачивается по 
мере прихода оригиналов.

Вычислите остаток каждого из менеджеров на 01.07.2021.

```python
balance = data.query('month == "июнь 2021" and document == "оригинал"')
balance = balance[(balance['receiving_date'].str.contains('2021-06') == False) & (balance['receiving_date'].str.contains('2021-05') == False)]

new_deals = balance[(balance['new/current'] == 'новая') & (balance['status'] == 'ОПЛАЧЕНО')]

current_deals_more_10k = balance[(balance['new/current'] == 'текущая') & (balance['status'] != 'ПРОСРОЧЕНО') & (balance['sum'] > 10000)]
current_deals_less_10k = balance[(balance['new/current'] == 'текущая') & (balance['status'] != 'ПРОСРОЧЕНО') & (balance['sum'] < 10000)]

new_deals['sum'] = new_deals['sum'].apply(lambda x: x * 0.07)
current_deals_more_10k['sum'] = current_deals_more_10k['sum'].apply(lambda x: x * 0.05)
current_deals_less_10k['sum'] = current_deals_more_10k['sum'].apply(lambda x: x * 0.03)

balance_01_07_2021 = pd.concat([new_deals, current_deals_more_10k, current_deals_less_10k], axis = 0)
balance_01_07_2021.groupby('sale')['sum'].sum()
```

Остаток каждого из менеджеров представлен в таблице выше
